**Code**


```
class Person:
  # Constructor: here the attributes are defined. this is also called initializer of the class.
  def __init__(self, name, age):
    self.name :str = name
    self.age :int = age
    
  def get_display(self) -> str:
    return f"<name :{self.name}, age:{self.age}>"
```

```
class Student(Person):
  def __init__(self, name, age, grade):
    Person.__init__(self, name, age) # inheriting the Person's attributes
    self.grade :int = grade

  
  def get_display_grade(self) -> int:
    return self.grade

  def get_display_student(self) -> str:
    return f"<name :{self.name}, age:{self.age}, grade: {self.grade}>"

Student1 = Student(name = "Kazuto", age = 17, grade = 7)
Student1.get_display_student()
  
```

```
class Int_to_roman:
  def romanconvert(self, num):
    value = [1000,900,500,400,100,90,50,40,10,9,5,4,1]
    symbol = ["M", "CM", "D", "CD", "C", "XC", "L","XL", "X", "IX", "V", "IV", "I"]
    Rnum = ''
    n = 0
    while num > 0:
      for i in range (num // value[n]):
        Rnum = Rnum + symbol[n]
        num = num - value[n]
      n += 1
    return Rnum 

print(Int_to_roman().romanconvert(1))
print(Int_to_roman().romanconvert(8))
print(Int_to_roman().romanconvert(120))
print(Int_to_roman().romanconvert(650))

```

```
class BankAccount:
    def __init__(self):
        self.balance = 0

    def withdraw(self, amount):
        self.balance -= amount
        return self.balance

    def deposit(self, amount):
        self.balance += amount
        return self.balance

class MinimumBalanceAccount(BankAccount):
    def __init__(self, minimum_balance):
        BankAccount.__init__(self)
        self.minimum_balance = minimum_balance

    def withdraw(self, amount):
        if self.balance - amount < self.minimum_balance:
            print('Sorry, minimum balance must be maintained.')
        else:
            BankAccount.withdraw(self, amount)
```

**Explaination** <br>
This code initially defines the super class "BankAccount", and the child class "MinimumBalanceAccount" that is inherited from the super class. In the super class, there is one initializer(constructor) of the class where the all attributes are defined, and two methods that allow user to withdraw and deposit money. Aside from that, MinimumBalanceAccount, the child class of "BankAccount" class, has one initializer that contains both minimum balance and the thing which inherites all BankAccount's attributes. At last, there is an if statement in the withdraw method so that users can ensure if minimum balance are maintained or not.
