**Culclate the tax**


```
def calculate_original_price(memory :int, screen_size :int, battery_capacity :int, storage_size :int) -> int:
  original_price = ((screen_size - memory)*storage_size ) * (battery_capacity/100)
  return original_price

def calculate_tax(year :int, original_price :int) -> int:
  if year < 2012:
    return original_price * 0.18
  elif year >= 2012:
    return original_price * 0.36

for mac in list_Macbooks:
  memory = mac.get_memory()
  screen_size = mac.get_screen_size()
  battery_capacity = mac.get_battery_capacity()
  storage_size= mac.get_storage_size()
  year = mac.get_year()

  original_price = calculate_original_price(memory, screen_size, battery_capacity, storage_size)
  tax = calculate_tax(original_price, year)


  print(f"{mac.get_pro_or_air()}-{mac.get_year()} pays TAX".ljust(40), f"{'.'*20}  {tax:.1f} dollars")  #:.1f means fix to 1 decimal figures


```

**Defining the class customer (example of polymorphism)**

```
class Customer:
  def __init__(self):
    self.macbooks = [] #empty list

  def add_macbook(self, pro_or_air, year, memory, screen_size, battery_capacity, storage_size, color, serial_number):
    #create a macbook
    new_macbook = Macbook(pro_or_air, year, memory, screen_size, battery_capacity, storage_size, color, serial_number)
    self.macbooks.append(new_macbook)

  def calculate_original_price (self):
    for mac in self.macbooks:
      memory = mac.get_memory()
      screen_size = mac.get_screen_size()
      battery_capacity = mac.get_battery_capacity()
      storage_size= mac.get_storage_size()
      pro_or_air = mac.get_pro_or_air()
      
      original_price = ((screen_size - memory)*storage_size ) * (battery_capacity/100)
      print("The price of {} is ... {} dollars".format(pro_or_air, original_price))

  def calculate_tax (self):
    for mac in self.macbooks:
        if year < 2012:
          print("The tax is ... {} dollars".format(original_price * 0.18))
        elif year >= 2012:
          print("The tax is ... {} dollars".format(original_price * 0.36))
```
