# **Unit3 - Quiz 13**

Given an array of food items and price, calculate the total price based on the table of tax categories below:

|      Item     | Tax Rate |
|:-------------:|:--------:|
|     Food:     |   10 %   |
| Electronics:  |    15%   |
|   Liqueur:    |    20%   |




```
list_food = ["Sushi", "Ramen", "Okonomiyaki"]
list_electronics = ["Bose speaker", "Nintendo Switch", "Macbook Pro"]
list_liquer = ["vodca", "corona beer", "wine"]

def final_price(order, original_price):
  total = 0
  length = len(order)
  #for i in range(length):
    #if order[i] in list_food:
      #total = total + original_price[i] * 1.1
    #if order[i] in list_electronics:
      #total = total + original_price[i] * 1.15
    #if order[i] in list_liquer:
      #total = total + original_price[i] * 1.2
  for item,price in zip(order,original_price):
    if item in list_food:
      total = total + price * 1.1
    if item in list_electronics:
      total = total + price * 1.15
    if item in list_liquer:
      total = total + price * 1.2

  return (total)

print(final_price(order = ["Sushi", "Bose speaker", "Vodca"], original_price = [600, 20000, 800]))
print(final_price(order = ["Ramen", "Nintendo Switch", "corona beer"], original_price = [1200, 50000, 500]))
print(final_price(order = ["Okonomiyaki", "Macbook Pro", "wine"], original_price = [1300, 150000, 2400]))
```
